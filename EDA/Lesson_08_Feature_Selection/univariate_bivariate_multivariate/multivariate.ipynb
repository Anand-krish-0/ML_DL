```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Generating synthetic data
np.random.seed(0)
var1 = np.random.normal(loc=10, scale=2, size=100)
var2 = np.random.normal(loc=20, scale=3, size=100)
var3 = np.random.normal(loc=30, scale=4, size=100)

# Create DataFrame
df = pd.DataFrame({'Var1': var1, 'Var2': var2, 'Var3': var3})

df
```

# 3D Scatter plot

A 3D scatter plot visualizes the relationship between three continuous variables by representing each data point in a three-dimensional space.

The position of each point is determined by the values of the three variables (Var1, Var2, Var3) along the x, y, and z axes, respectively.

It allows for the exploration of patterns, clusters, and trends within the data across multiple dimensions.

```python

# 3D Scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Var1'], df['Var2'], df['Var3'], c='blue', marker='o')
ax.set_xlabel('Var1')
ax.set_ylabel('Var2')
ax.set_zlabel('Var3')
ax.set_title('3D Scatter Plot of Var1, Var2, and Var3')
plt.show()
```

# Bubble plot

A bubble plot is similar to a scatter plot but with an additional dimension represented by the size of the markers (bubbles).

In the context of three continuous variables (Var1, Var2, Var3), the positions of the data points on the x and y axes represent the values of Var1 and Var2, while the size of each bubble represents the values of Var3.

Bubble plots are useful for visualizing the relationship between two variables while incorporating the magnitude of a third variable.

```python
# Bubble plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Var1'], df['Var2'], s=df['Var3']*10, alpha=0.5)
plt.xlabel('Var1')
plt.ylabel('Var2')
plt.title('Bubble Plot of Var1 vs. Var2 (Var3 as Bubble Size)')
plt.grid(True)
plt.show()
```

# Pair plot

A pair plot shows pairwise relationships between multiple variables in a grid of scatter plots.

For three variables (Var1, Var2, Var3), a pair plot would display scatter plots of Var1 vs. Var2, Var1 vs. Var3, and Var2 vs. Var3.

Pair plots provide a quick way to visualize correlations, patterns, and potential dependencies between variables in a multivariate dataset.

```python
# Pair plot
sns.pairplot(df)
plt.suptitle('Pairwise Relationships between Variables')
plt.show()
```

# Parallel coordinates plot

A parallel coordinates plot displays multivariate data by plotting each variable on a separate axis and connecting points across axes to show relationships.

Each data point is represented by a line segment that connects its values across the axes corresponding to different variables (Var1, Var2, Var3).

Parallel coordinates plots are useful for identifying patterns, clusters, and trends in multivariate datasets, particularly when dealing with high-dimensional data.

```python
# Parallel coordinates plot
plt.figure(figsize=(10, 8))
pd.plotting.parallel_coordinates(df, 'Var3', colormap='viridis')
plt.title('Parallel Coordinates Plot of Variables')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.show()
```

# Radar chart

A radar chart represents the mean values of variables (Var1, Var2, Var3) on a two-dimensional chart with multiple axes radiating from a central point.

Each variable is plotted along its own axis, and the values of the variables are represented by the distance of the data point from the center along each axis.

Radar charts are commonly used to compare the overall profiles or characteristics of multiple variables across different categories or groups.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample DataFrame
df = pd.DataFrame({
    'Variable': ['Var1', 'Var2', 'Var3'],
    'Value': [10, 20, 30]  # Example mean values for demonstration
})

# Number of variables
categories = list(df['Variable'])
N = len(categories)

# Create angles for radar chart
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()

# Make the plot close entirely
values = df['Value'].tolist()
values += values[:1]
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2, linestyle='solid')
ax.set_yticklabels([])

# Add labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

plt.title('Radar Chart of Mean Values of Variables')
plt.show()
```
