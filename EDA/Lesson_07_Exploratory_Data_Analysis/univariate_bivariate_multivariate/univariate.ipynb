```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample age data
data = {
    'Age': [25, 30, 22, 45, 28, 35, 40, 18, 50, 27]
}

# Convert data to a DataFrame
df = pd.DataFrame(data)
```

# Histograms

Histograms are useful for visualizing the distribution of a single variable, such as age in this case. It divides the data into bins and represents the frequency or proportion of observations in each bin.

In the age dataset, a histogram would show the frequency of individuals falling into different age ranges (e.g., 18-25, 26-35, 36-45, etc.).

```python
# 1. Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# Box plot

Box plots provide a visual summary of the central tendency, dispersion, and skewness of the data, as well as identify potential outliers.

For the age dataset, a box plot would display the minimum, first quartile, median, third quartile, and maximum age, as well as any outliers.

```python
# 2. Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

# Kernel density plots

Kernel density plots estimate the probability density function of a continuous variable.

In the age dataset, a kernel density plot would show the distribution of ages as a smooth curve, providing insights into the shape of the distribution.

```python
# 3. Kernel density plot
plt.figure(figsize=(8, 6))
df['Age'].plot(kind='kde', color='orange')
plt.title('Kernel Density Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

# Bar charts

Bar charts are suitable for displaying the frequency or proportion of different categories in categorical data.

If we categorize age into ranges (e.g., 0-10, 11-20, 21-30, etc.), a bar chart would show the frequency of individuals in each age range.

```python
# 4. Bar chart
age_ranges = ['0-20', '21-30', '31-40', '41-50', '51+']
age_bins = [0, 20, 30, 40, 50, float('inf')]
df['Age Range'] = pd.cut(df['Age'], bins=age_bins, labels=age_ranges)
age_counts = df['Age Range'].value_counts()

plt.figure(figsize=(8, 6))
age_counts.plot(kind='bar', color='green')
plt.title('Bar Chart of Age Ranges')
plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# Pie charts

Pie charts show the composition of categorical data as slices of a pie, with each slice representing a category's proportion.

In the age dataset, if we categorize age groups into, say, young (0-25), middle-aged (26-50), and elderly (51 and above), a pie chart would display the proportion of individuals in each age group.

```python
# 5. Pie chart
plt.figure(figsize=(8, 6))
age_counts.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink'])
plt.title('Pie Chart of Age Ranges')
plt.ylabel('')
plt.show()
```
